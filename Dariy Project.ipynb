```python
#from the dataset mentioned above, find the answers of these few questions 
# Q1. does number of cows in a farm has any dependency on area aquired. 
#Ans--number of cows and area aquired by them has propertional relationship.
```

```python
gb = df.groupby("Farm Size").agg({"Number of Cows":'sum',"Total Land Area (acres)":'sum'})
gb
```

```python
# import datetime
# df["Production Date"] = df["Production Date"].astype('datetime64[ns]')
# df["Expiration Date"] = df["Expiration Date"].astype('datetime64[ns]')
```

```python
df["Production Date"] = pd.to_datetime(df['Production Date'])
df["Expiration Date"] = pd.to_datetime(df["Expiration Date"])
```

```python
x= gb.corr()
x
```

```python
df["Shelf Life"] = (abs(df["Production Date"] - df["Expiration Date"]))
df.head()
```

```python
#2. what is the average expiry days of each product. 

gb = df.groupby("Product Name").agg({'Shelf Life': 'mean'})
gb
```

```python
# Q3. mention the name of the products with max and min shelf life
gb1 = df.groupby("Product Name").agg({'Shelf Life': 'max'})
gb2 = df.groupby("Product Name").agg({'Shelf Life': 'min'})
gb2
x=pd.merge(left = gb1, right = gb2, on = "Product Name", how = "inner")
x
```

```python
# 4. summarize the data on the basis of brands and land acquired by them,cows and the revenue generated by them. 
gb = df.groupby("Brand").agg({"Total Land Area (acres)":'sum','Number of Cows': 'sum',"Approx. Total Revenue(INR)": 'sum'})
gb
```

```python
#Q 5. What is the average shelf life for different storage conditions. 
gb = df.groupby("Storage Condition").agg({'Shelf Life': 'mean'})
gb
```
